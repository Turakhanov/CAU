### Tasks 1

Implement:  
• Merge Sort  
• Quick Sort  
• Binary Search  
• Linear Search (for comparison)  
• Add depth: 
Print intermediate steps (recursion tree)  
Count:  
• comparisons  
• swaps  
Compare runtime using time module

```python
import time
import random

class Counter:
    def __init__(self):
        self.comparisons = 0
        self.swaps = 0

def merge_sort(arr, depth=0, counter=None):
    if counter is None:
        counter = Counter()

    print("  " * depth + f"Split: {arr}")

    if len(arr) <= 1:
        return arr, counter

    mid = len(arr) // 2
    left, counter = merge_sort(arr[:mid], depth + 1, counter)
    right, counter = merge_sort(arr[mid:], depth + 1, counter)

    merged = []
    i = j = 0

    while i < len(left) and j < len(right):
        counter.comparisons += 1
        if left[i] < right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1
            counter.swaps += 1

    merged.extend(left[i:])
    merged.extend(right[j:])

    print("  " * depth + f"Merge: {merged}")
    return merged, counter


def quick_sort(arr, depth=0, counter=None):
    if counter is None:
        counter = Counter()

    print("  " * depth + f"QuickSort: {arr}")

    if len(arr) <= 1:
        return arr, counter

    pivot = arr[len(arr) // 2]
    left, middle, right = [], [], []

    for x in arr:
        counter.comparisons += 1
        if x < pivot:
            left.append(x)
        elif x > pivot:
            right.append(x)
            counter.swaps += 1
        else:
            middle.append(x)

    l_sorted, counter = quick_sort(left, depth + 1, counter)
    r_sorted, counter = quick_sort(right, depth + 1, counter)

    result = l_sorted + middle + r_sorted
    print("  " * depth + f"Result: {result}")
    return result, counter


def linear_search(arr, target):
    comparisons = 0
    for i, val in enumerate(arr):
        comparisons += 1
        if val == target:
            return i, comparisons
    return -1, comparisons


def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    comparisons = 0

    while left <= right:
        mid = (left + right) // 2
        comparisons += 1

        if arr[mid] == target:
            return mid, comparisons
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1, comparisons


arr = [random.randint(1, 100) for _ in range(10)]
target = arr[5]

print("\nOriginal Array:", arr)

start = time.time()
sorted_merge, merge_counter = merge_sort(arr.copy())
merge_time = time.time() - start

print("\nMerge Sort Result:", sorted_merge)
print("Comparisons:", merge_counter.comparisons)
print("Swaps:", merge_counter.swaps)
print("Time:", merge_time)

start = time.time()
sorted_quick, quick_counter = quick_sort(arr.copy())
quick_time = time.time() - start

print("\nQuick Sort Result:", sorted_quick)
print("Comparisons:", quick_counter.comparisons)
print("Swaps:", quick_counter.swaps)
print("Time:", quick_time)

start = time.time()
idx, comp = linear_search(arr, target)
linear_time = time.time() - start

print("\nLinear Search:")
print("Found at index:", idx)
print("Comparisons:", comp)
print("Time:", linear_time)

start = time.time()
idx, comp = binary_search(sorted_merge, target)
binary_time = time.time() - start

print("\nBinary Search:")
print("Found at index:", idx)
print("Comparisons:", comp)
print("Time:", binary_time)
```

### Task 2

You are given: 
1 A string text of length n  
2 A string pattern of length m  
Your task is to: 
1. Preprocess the pattern to compute the LPS (Longest Prefix Suffix) array  
2. Use the LPS array to search for the pattern inside the text  
3. Print all starting indices where the pattern is found  
Requirements 
1. Implement two functions:  
• compute_lps(pattern)  
• kmp_search(text, pattern)  
2. The program should:  
• Avoid re-checking characters unnecessarily  
• Use previously computed information (LPS array)  
3. Output example:  
Match found at index 10 

```python
def compute_lps(pattern):
    m = len(pattern)
    lps = [0] * m

    length = 0 
    i = 1

    while i < m:
        if pattern[i] == pattern[length]:
            length += 1
            lps[i] = length
            i += 1
        else:
            if length != 0:
                length = lps[length - 1] 
            else:
                lps[i] = 0
                i += 1

    return lps


def kmp_search(text, pattern):
    n = len(text)
    m = len(pattern)

    lps = compute_lps(pattern)

    i = 0 
    j = 0 

    while i < n:
        if text[i] == pattern[j]:
            i += 1
            j += 1

        if j == m:
            print(f"Match found at index {i - j}")
            j = lps[j - 1] 

        elif i < n and text[i] != pattern[j]:
            if j != 0:
                j = lps[j - 1] 
            else:
                i += 1



text = "ababcabcabababd"
pattern = "ababd"

kmp_search(text, pattern)
```

### Task 3

3. Labyrinth 
• Time limit: 1.00 s 
• Memory limit: 512 MB 
You are given a map of a labyrinth, and your task is to find a path from start to end. You can 
walk left, right, up and down. 
Input 
The first input line has two integers nnn and mmm: the height and width of the map. 
Then there are nnn lines of mmm characters describing the labyrinth. Each character is . (floor), # 
(wall), A (start), or B (end). There is exactly one A and one B in the input. 
Output 
First print "YES", if there is a path, and "NO" otherwise. 
If there is a path, print the length of the shortest such path and its description as a string 
consisting of characters L (left), R (right), U (up), and D (down). You can print any valid 
solution. 
Constraints 
• 1≤n,m≤10001 \le n,m \le 10001≤n,m≤1000 
Example 
Input: 
5 8 
######## 
#.A#...# 
#.##.#B# 
#......# 
######## 
Output: 
YES 
9 
LDDRRRRRU

```python
from collections import deque

def solve():
    n, m = map(int, input().split())
    grid = [list(input()) for _ in range(n)]

    
    directions = [
        (-1, 0, 'U'),
        (1, 0, 'D'),
        (0, -1, 'L'),
        (0, 1, 'R')
    ]

    for i in range(n):
        for j in range(m):
            if grid[i][j] == 'A':
                start = (i, j)

    queue = deque([start])
    visited = [[False]*m for _ in range(n)]
    visited[start[0]][start[1]] = True


    parent = [[None]*m for _ in range(n)]

    while queue:
        x, y = queue.popleft()

        if grid[x][y] == 'B':
            # reconstruct path
            path = []
            while (x, y) != start:
                px, py, move = parent[x][y]
                path.append(move)
                x, y = px, py
            path.reverse()

            print("YES")
            print(len(path))
            print("".join(path))
            return

        for dx, dy, move in directions:
            nx, ny = x + dx, y + dy

            if 0 <= nx < n and 0 <= ny < m:
                if not visited[nx][ny] and grid[nx][ny] != '#':
                    visited[nx][ny] = True
                    parent[nx][ny] = (x, y, move)
                    queue.append((nx, ny))

    print("NO")


solve()
    
```
